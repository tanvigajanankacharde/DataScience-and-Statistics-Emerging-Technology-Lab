```python
#import  library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

# Line chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.show()
```

```python
x=np.random.randint(1,10,9)
x
```

```python
y=np.random.randint(1,10,9)
y
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")

plt.ylabel("y axis")

plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.bar(x,y, color="purple")
plt.show()
```

```python
 a=(1,5,4,7,6,9,3,7,1,4,6,9,1)
 b=(10,20,12,51,42,47,43,32,20,29,25,78,10) 
plt.scatter(a,b)
plt.title("scatter Chart")
plt.xlabel("x axis")

plt.ylabel("y axis")
plt.scatter(a,b, color="red")
plt.show()
```
